```python
#W7A1
def binary_search(a,k):
    left = 0
    right = len(a)-1
    while left <= right:
        mid = (left + right)//2
        if a[mid] == k:
            return mid 
        elif a[mid] < k:
            left = mid+1
        else:
            right = mid-1
    return -1
a = list(map(int,input().split()))
k = int(input())
print(binary_search(a,k))
```

```python
#W7A2
def solanxh(a,k):
    cnt = 0
    for i in range(len(a)):
        if a[i] == k:
            cnt += 1
    return cnt
a = list(map(int,input().split()))
k = int(input())
print(solanxh(a,k))
    
```

```python
#W7A3
def bulle_sort(a):
    for i in range(len(a)-1):
        for j in range(len(a)-1-i):
            if a[j] > a[j+1]:
                a[j], a[j+1] = a[j+1], a[j]
    return a
n = list(map(int, input().split()))
print(bulle_sort(n))
```

```python
#W7A4
def xhnhieunhat(a):
    dem = {}
    max_cnt = 0
    ket_qua = 0
    for x in a:
        dem[x] = dem.get(x, 0) +1
        if dem[x] > max_cnt:
            max_cnt = dem[x]
            ket_qua = x
    return ket_qua, max_cnt
a = list(map(int,input().split()))
print(xhnhieunhat(a))

            
```

```python
#W7A5
def tongcapso(a,x):
    cnt = 0
    for i in range(len(a)):
        for j in range(i+1,len(a)):
            if a[i] + a[j] == x:
                cnt +=1
    return cnt
a = list(map(int, input().split()))
x = int(input())
print(tongcapso(a,x))
                
        
```

```python
#W7A7
def len_day_con_tang(a):
    if not a:
        return 0
    n = len(a)
    dp = [1] * n
    for i in range(n):
        for j in range(i):
            if a[j] < a[i]:
                dp[i] = max(dp[i],dp[j] +1)
    return max(dp)
a = list(map(int,input().split()))
print(len_day_con_tang(a))
```

```python
#w7A9
import bisect
def dem_so_nho_hon(a):
    lst = []
    kqua = []
    for x in reversed(a):
        pos = bisect.bisect_left(lst, x)
        kqua.append(pos)
        lst.insert(pos, x)
    return kqua[::-1]
a = list(map(int,input().split()))
print(dem_so_nho_hon(a))
        
```

```python
#W7A10
def max_lst(a, m):
    n = len(a)
    b = 0
    for i in range(n):
        s = 0
        for j in range(i,n):
            s += a[j]
            c = s % m
            if c > b:
                b = c
    return b
a = list(map(int,input().split()))
m = int(input())
print(max_lst(a,m))


```
