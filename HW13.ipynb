```python
#W11A1
def reverse(a:str) -> int:
    with open(a,"r",encoding="utf-8") as f:
        n = f.read().strip()
    reverse_n = n[::-1]
    return int(reverse_n)
print(reverse("so.txt"))
            
    
```

```python
#W11A2
def canWinNim(a: str) -> bool:
    with open(a, "r", encoding="utf-8") as f:
        n = int(f.read().strip())
    return n % 4 != 0
print(canWinNim("num.txt"))
```

```python
#W11A5
def count(path:str, charater:str):
    with open(path,"r",encoding="utf-8") as f:
        sentence = f.read().strip()
    return sentence.count(charater)
print(count("xau.txt","n"))
        
```

```python
#W11A8
n = int(input("Nhập n (n ≤ 20): "))
f = [""] * (n + 1)
if n >= 0:
    f[0] = "a"
if n >= 1:
    f[1] = "b"
for i in range(2, n + 1):
    f[i] = f[i - 1] + f[i - 2]
with open("output.txt", "w", encoding="utf-8") as file:
    for i in range(n + 1):
        file.write(f"f({i}) = {f[i]}\n")
print("Đã ghi các từ Fibonacci vào file output.txt")
```

```python
#W11A9
def grade10(path: str):
    result = []
    with open(path, "r", encoding="utf-8") as f:
        lines = f.readlines()
    for line in lines[1:]:  # bỏ dòng tiêu đề
        data = line.strip().split(",")
        ho = data[1]
        ten = data[2]
        thcs4 = data[5]
        if thcs4 == "10":
            result.append(ho + " " + ten)
    return result
print(grade10("chuoi1.txt"))
```
