```python
#W6A5
def dem(a):
        d={'positives':0,'negatives':0,'zeros':0}
        for i in a:
            if i>0:
                d['positives'] +=1
            elif i<0:
                d['negatives'] +=1
            else:
                d['zeros'] +=1
        return d
n = list(map(int, input().split()))
print(dem(n))
```

```python
#W6A9 Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới
#trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có
#key. 
d1 = input().split()
dict1 = {}
for i in range(0, len(d1), 2):
    dict1[d1[i]] = int(d1[i+1])
d2 = input().split()
dict2 = {}
for i in range(0, len(d2), 2):
    dict2[d2[i]] = int(d2[i+1])
result = {}
for k, v in dict1.items():
    result[k] = v
for k, v in dict2.items():
    result[k] = result.get(k, 0) + v
sorted_result = dict(sorted(result.items()))
print(sorted_result)

    
    
    


    
    


```

```python
#W6A10
a = list(map(int,input().split()))
k = int(input())
lst = []
for i in range(len(a)):
    for j in range(i+1,len(a)):
        if a[i]+a[j]==k:
            lst.append((a[i],a[j]))
print(lst)
            


```

```python
#W6A12
a = list(map(int, input().split()))
max_count = 0
ans = a[0]
for x in a:
    c = a.count(x)
    if c > max_count or (c == max_count and x < ans):
        max_count = c
        ans = x
print(ans)
```
