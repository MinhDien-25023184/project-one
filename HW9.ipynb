```python
#W7A16
def kiemtramang(a):
    dict = {}
    for x in a:
        if x in dict:
            dict[x] += 1
        else:
            dict[x] = 1
    remove = 0
    for x in dict:
        cnt = dict[x]
        if cnt > x:
            remove += cnt - x
        elif cnt < x:
            remove = cnt
    if remove == 0:
        return 'YES'
    else:
        return 'NO', remove
a = list(map(int,input().split()))
print(kiemtramang(a))
    
```

```python
#W7A18
def sxep_chan_le(a):
    chan = []
    le = []
    for x in a:
        if x%2==0:
            chan.append(x)
        else:
            le.append(x)
    for i in range(len(chan)):
        for j in range(i,len(chan)):
            if chan[j] <chan[i]:
                chan[i],chan[j] = chan[j],chan[i]
    for x in range(len(le)):
        for y in range(x,len(le)):
            if le[x] < le[y]:
                le[x],le[y] = le[y],le[x]
    lst = chan + le
    return lst
a = list(map(int,input().split()))
print(sxep_chan_le(a))
    
                
                
```

```python
#W7A20
def tong_so_dau(a,k):
    sum = 0
    lst = []
    for x in a:
        n = abs(x)
        cnt = 0
        while n > 0:
            cnt +=1
            n //= 10
        if cnt == k:
            lst.append(x)
    for i in lst:
        sum += i
    return sum
a = list(map(int,input().split()))
k = int(input())
print(tong_so_dau(a,k))
    
    
```

```python
#W7A21
def tim_so_ngay_cua_nam(d,m):
    day = [31,29,31,30,31,30,31,31,30,31,30,31]
    if m<1 or m > 12:
        return "invalid"
    if d<1 or d>day[m-1]:
        return "invalid"
    tong = sum(day[:m-1])+d
    return tong
d,m = map(int, input().split())
print(tim_so_ngay_cua_nam(d,m))
```

```python
#W7A22
def ten_hay_gap_nhat(a):
    cnt ={}
    for full_name in a:
        name = full_name.strip().split()[-1]
        if name in cnt:
            cnt[name] += 1
        else:
            cnt[name] = 1
    max_cnt = 0
    for name in cnt:
        if cnt[name] > max_cnt:
            max_cnt = cnt[name]
    result = None
    for name in cnt:
        if cnt[name] == max_cnt:
            if result is None or name < result:
                result = name

    return result
a = input().split()
print(ten_hay_gap_nhat(a))
```
