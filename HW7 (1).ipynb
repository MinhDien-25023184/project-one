```python
#W6A24
def move_zero_front(lst):
    zeros = lst.count(0)
    return [0]*zeros + [i for i in lst if i != 0]

def move_zero_end(lst):
    zeros = lst.count(0)
    return [i for i in lst if i != 0] + [0]*zeros
```

```python
#W6A26
def local_max_sum(lst):
    s = 0
    for i in range(1, len(lst)-1):
        if lst[i] > lst[i-1] and lst[i] > lst[i+1]:
            s += lst[i]
    return s
```

```python
#W6A28
def check_sequence(a):
    if all(a[i] < a[i+1] for i in range(len(a)-1)):
        return "tăng thực sự"
    if all(a[i] > a[i+1] for i in range(len(a)-1)):
        return "giảm thực sự"
    return "vô danh"
```
