```python
# Viết chương trình tính giai thừa của một số nguyên n cho trước (n!=1×2×…×n)
# với 0 < n < 100. 
# W4A3
try:
    n=int(input())
    giaithua=1
    for i in range (1,n+1):
        giaithua=giaithua*i
    print(giaithua)
except ValueError:
    print("Nhập dữ liệu sai")
```

```python
#Bài7
import math
n=0
A = []
def kiemtranguyento(n):
    for i in range(2,int(math.sqrt(n)+1)):
         if n%i == 0:
             for a in range(2,int(math.sqrt(i)+1)):
                 if i%a==0:
                    break
                 else:
                     A.append(i)
         else:
             A.append(n)
    print(max(A))
print(kiemtranguyento(9))
        
            

```

```python
#bai10
n = int(input())
max_dem = 0   
a_value = 1 
for a in range(1,n+1):
    b=a
    count = 0
    while b != 1:
        if b%2==0:
           b=b//2
        else:
           b=b*3+1
        count += 1
    if count > max_dem:
        max_dem = count
        a_value = a
    elif count == max_dem:
        a_value = min(a_value, a)  
print(a_value, max_dem)
```
