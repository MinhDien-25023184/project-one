```python
#W5A1
def Max_of_two(a,b):
    if a==b:
        return False
    else:
        if a>b:
            return a
        else:
            return b
try:
    a,b=map(int, input().split())
    print(Max_of_two(a,b)) 
except ValueError:
    print("Bạn đã nhập sai dữ liệu")
```

```python
#W5A2
def swap(a,b):
    return b, a
print(swap(3,7))
```

```python
#W5A4
def sohoanhao(n):
    if n <= 0:
        return False
    
    tong = 0
    for i in range(1, n):
        if n % i == 0:
            tong += i
    
    return tong == n
print(sohoanhao(6))
```

```python
#W5A3
def kiemtrasonguyento(n):
    if n <= 1:
        return False
    else:
        for i in range(2,n):
            if n%i==0:
                return False
            else:
                return True
print(kiemtrasonguyento(7))
```

```python
#W5A5
def timvitriso(lst,k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1
a=(1,2,3,4,5,6)
print(timvitriso(a,5))
```

```python
#W5A6
def giaithua(n):
    if n < 0:
        return 0
    
    gt = 1
    for i in range(1, n + 1):
        gt *= i

    return gt
print(giaithua(5))
```

```python
#W5A7
def calculator(so1,so2,pheptinh):
    if pheptinh == '+':
        kqua = so1 + so2
    elif pheptinh == '-':
        kqua = so1 - so2
    elif pheptinh ==  '*':
        kqua = so1 * so2
    elif pheptinh == '/':
        if so2 == 0:
            return None
        else:
            kqua = so1 / so2
    return kqua
print(calculator(1,2,'*'))
         
```

```python
#W5A9
def tongcacchuso(n):
    tong=0
    while n !=0:
        tong += n%10
        n = (n-n%10)//10
    return tong
n=int(input())
print(tongcacchuso(n))
        
```

```python
#W5A8
def hamming_distance(x, y):
    return bin(x ^ y).count("1")
print(hamming_distance(3,5))
```

```python
#W5A10
def is_isomorphic(a, b):
    map_a_to_b = {}
    map_b_to_a = {}

    for x, y in zip(a, b):
        if x in map_a_to_b:
            if map_a_to_b[x] != y:
                return False
        else:
            map_a_to_b[x] = y
        if y in map_b_to_a:
            if map_b_to_a[y] != x:
                return False
        else:
            map_b_to_a[y] = x

    return True
```
