```python
# bài3
a=int(input("Nhập vào số nguyên dương a:"))
print(f"Bảng cửu chương của {a} là:")
for i in range(1,11):
    x = a*i
    print(x,end=' ')
```

```python
# Bài4
h=int(input("nhập chiều cao của tam giác"))
for i in range(1,h+1):
    print(" " * (h - i),"*" * (2 * i - 1))
```

```python
# Bài7
a=int(input("Nhập vào số nguyên dương:"))
cnt=0
for i in range(1,a+1):
    if a%i==0:
        cnt +=1
print(cnt)
```

```python
# Bài 8
a,b=map(int, input("Nhập vào 2 số nguyên dương:").split())
print(f"Ước chung của {a} và {b} là:")
for i in range(1,min(a,b)+1):
    if a%i==0 and b%i==0:
        print(i,end =' ')
```

```python
# Bài 9
import math
a = int(input("Nhập vào số nguyên dương a:"))
i = int(math.sqrt(a))
if i*i==a:
    print(f"{a} là số chính phương")
else:
    print(f"{a} ko là số chính phương")
```
