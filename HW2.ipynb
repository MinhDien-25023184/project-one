```python
#BAI3
import math
a,b,c=map(float, input("Độ dài các cách:").split())
if a+b<c or b+c<a or c+a<b:
    print("Không phải 3 cạnh 1 tam giác")
else:
    if a==b==c:
       loai="tam giác đều"
    elif a==b or b==c or c==a:
       loai="tam giác cân"
    elif a**2==b**2+c**2 or b**2==a**2+c**2 or c**2==a**2+b**2:
       loai="tam giác vuông"
    else:
       loai="tam giác thường"
P=a+b+c
p=P/2
S= math.sqrt(p*(p-a)*(p-b)*(p-c))
print(loai)
print("Diện tích=", S)
print("Chu vi=", P)
```
